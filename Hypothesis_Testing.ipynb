#Hypothesis Testing

1.

   - The null hypothesis is a statement that assumes no effect, no difference, or no relationship in the population. It is important because hypothesis testing is designed to evaluate whether there is enough statistical evidence to refect the null hypothesis in favor of an alternative hypothesis.

2.

   - The significance level represents the probaiblity of making a type l error, which means rejecting a true nulll hypothesis. Common values of a are 0.05 or 0.01, indication a 5% or 1% risk of making such an error.

3.

   - Type | error: Rejecting the null hypothesis when it is actually true.

   - Type || error: Failing to reject the null hypothesis when it is actually false.

4.

   - A one-tailed test check sfor an deffect in one direction only. Example: Testing whether the average score is greater than 70.

   - A two-tailed test checks for an effect in both directions.     Example: Testing whether the average score is different from 70.

6.

   - A Z-test should be used when:

   - The population standard deviation is known, and

   - The sample size is large (n>30).

   - If the population standard deviation is unknown and the sample size is small, a t-test is preferred.

7.

   - Productivity before and after training was measured for the same 6 employee.

   - Hypothesis:

   - H0: There is no improvement in productivity after training.

   - H1: Productivity has improved after training.

   - A paired t-test is appropriate because measurements are taken from the same individuals.

   - Conclusion:

   - At a = 0.05, the null hypothesis is rejected. There is sufficient evidence to conclude that the training program improved employee productivity.

8.

   - Given:

   - A contingency table of gender and product preference is provided.

   - Hypotheses:

   - H0: Product preference is independent of gender.

   - H1: Product preference is not independent of gender.

   - A Chi-square test of independence is used.

   - Conclusion:

   - At a = 0.05, the null hypothesis is rejected.

   - There is a significant relationship between gender and product preference.
